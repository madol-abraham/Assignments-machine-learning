```python
# YOUR CODE HERE
plt.hist(df['city'],edgecolor='black')
plt.title('city fuel Economy')
plt.xlabel('city')
plt.ylabel('frequency')
plt.show()
```

The histogram displays the distribution of city fuel economy . The data follows a roughly normal distribution, centered around 20 , with most vehicles clustering within the 15-25 range. The frequency of vehicles with lower or higher fuel efficiency decreases as you move away from the center. The graph shows that a majority of vehicles have a city fuel economy close to the average, with fewer vehicles achieving very high or very low fuel economy.

# Correlation Heatmap

```python
df_temp = df.select_dtypes(include=['float64','int64'])

corr_matric = df_temp.corr()
sns.heatmap(corr_matric,annot=False,cmap='coolwarm')
plt.show()
```

The diagram suggests that engine displacement and the number of cylinders appear to have an inverse relationship with both the fuel efficiency score  and the greenhouse gas score . This implies that as the size of the engine and the number of cylinders increase, the vehicle tends to become less efficient in terms of these scores.

Additionally, it highlights that engine displacement and cylinders are positively correlated with CO2 emissions, which means that larger engines tend to produce higher emissions.
